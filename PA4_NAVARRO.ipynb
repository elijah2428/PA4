```python
# Import the pandas library, used for data manipulation and analysis.
import pandas as pd
# Import the pyplot module from matplotlib for creating visualizations (e.g., plots, charts).
import matplotlib.pyplot as plt
```

```python
# Load the Excel file 'board2.xlsx' into a DataFrame called 'board'
board = pd.read_excel('board2.xlsx')
board
```

```python
# Filter the 'board' DataFrame for the following conditions:
# 1. 'Hometown' is 'Luzon'
# 2. 'Track' is 'Instrumentation'
# 3. 'Electronics' score is greater than 70
# Select only the 'Name', 'GEAS', and 'Electronics' columns from the rows that meet the conditions
instru = board.loc[(board['Hometown'] == 'Luzon')&
          (board['Track'] == 'Instrumentation')&
          (board['Electronics']>70), ['Name', 'GEAS', 'Electronics']]
```

```python
instru
```

```python
#This calculates the row-wise average of the columns 'Math', 'Electronics', 'GEAS', and 'Communication'
board['Average'] = board[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1)
# Filter the 'board' DataFrame for the following conditions:
# 1. 'Hometown' is 'Mindanao'
# 2. 'Gender' is 'Female'
# 3. 'Average' score is greater than or equal to 55
# Select only the 'Name', 'Track', 'Electronics', and 'Average' columns from the rows that meet the conditions
mindy = board.loc[(board['Hometown'] == 'Mindanao')&
                  (board['Gender'] == 'Female')&
                  (board['Average']>=55), ['Name', 'Track', 'Electronics', 'Average']]
```

```python
mindy
```

```python
plt.figure(figsize=(15, 10))

# 1. Boxplot for Average Grade by Track
# Removed plt.subplot and let boxplot handle the layout
board.boxplot(column='Average_Grade', by='Track', grid=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Average Grade by Track')
plt.suptitle('')  # Remove the default title added by boxplot
plt.xlabel('Track')
plt.ylabel('Average Grade')

# 2. Boxplot for Average Grade by Gender
plt.figure(figsize=(15, 10))  # Create a new figure for the next boxplot
board.boxplot(column='Average_Grade', by='Gender', grid=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Average Grade by Gender')
plt.suptitle('')  # Remove the default title added by boxplot
plt.xlabel('Gender')
plt.ylabel('Average Grade')

# 3. Boxplot for Average Grade by Hometown
plt.figure(figsize=(15, 10))  # Create a new figure for the next boxplot
board.boxplot(column='Average_Grade', by='Hometown', grid=False, patch_artist=True, boxprops=dict(facecolor='lightcoral'))
plt.title('Average Grade by Hometown')
plt.suptitle('')  # Remove the default title added by boxplot
plt.xlabel('Hometown')
plt.ylabel('Average Grade')

# Show the plots
plt.show()
```
